```python
sns.set()
rng = np.random.RandomState(1)
X = np.dot(rng.rand(2, 2), rng.randn(2, 200)).T
plt.scatter(X[:, 0], X[:, 1]);
```

<div style="direction:rtl;line-height:300%">
    <font face="XB Zar" size=4>
متد PCA به کمک یافتن 
        محورهایی به نام Principal Components یا PCs نگاشتی از داده‌های ورودی (x)
        بر روی این محورها ایجاد می‌کند.
        در اینجا از تابع آماده‌ی کتابخانه‌ی sklearn و با استفاده از ۲ محور PC تبدیل را انجام می‌دهیم.
        در این حالت چون داده‌هایمان ۲ بعدی هستند کاهش ابعادی رخ نمی‌دهد:
    </font>

```python
# Performing PCA
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
pca.fit(nparr)
f = pca.transform(nparr)
# PC vectors (pivots) and eigen values [of covariance matrix of data]
print(pca.components_)
print(pca.explained_variance_)
```

```python
f.shape
```

```python
# Visualizing data including PC pivots
def draw_vector(v0, v1, ax=None):
    ax = ax or plt.gca()
    arrowprops=dict(arrowstyle='->',
                    linewidth=4,
                    shrinkA=0, shrinkB=0, color='r')
    ax.annotate('', v1, v0, arrowprops=arrowprops)

# plot data
plt.scatter(X[:, 0], X[:, 1], alpha=0.2)
for length, vector in zip(pca.explained_variance_, pca.components_):
    v = vector * 3 * np.sqrt(length)
    draw_vector(pca.mean_, pca.mean_ + v)
```

در مرحله اول، باید با یافتن میانگین داده‌ها، آن‌ها را به مرکز مختصات ببرید. نمونه‌ی این نگاشت در فضای ۲ بعدی در  تصاویر زیر آمده است:

```python
# Generate a matrix of values having mean of 0
def normalization_by_only_mean(data):
    mean = data.mean(axis=0)
    std = data.std(axis=0)
    normalized_data = (data - data.mean(axis=0))
    B = normalized_data
    return B, mean, std

def normalization_by_std_mean(data):
    mean = data.mean(axis=0)
    std = data.std(axis=0)
    normalized_data = (data - mean)/std
    B = normalized_data
    return B, mean, std

B, mean, std = normalization_by_only_mean(nparr)
B.mean(axis=0)
# Your code here
```

<div dir="rtl", style="line-height:300%;">
<font face="XB Zar" size=4>
حال می‌توان به سراغ یافتن ماتریس <strong>کوواریانس</strong> رفت. ابتدا توضیحاتی در مورد این ماتریس ارائه می‌‌‌‌کنیم و سپس روش به دست آوردن آن را بیان می‌‌‌‌نماییم.
اگر داده‌‌‌‌های ما n تایی باشند، ماتریس کوواریانس، یک ماتریس n*n با (به عبارت دیگر هر بردار داده عضو فضای n بعدی R^nاست. بنابراین در تعریف ماتریس کوواریانس  درایه ijام آن کواریانس i ام و jام درایه بردار داده است)ویژگی‌‌‌‌های زیر است:
<br>
-عناصر موجود بر روی قطر اصلی، نشان‌‌‌‌دهنده‌‌‌‌ی واریانس داده‌‌‌‌های هر بعد هستند.
<br>
-عناصر غیر قطر اصلی، نشان‌‌‌‌دهنده‌‌‌‌ی کوواریانس دو به دوی داده‌‌‌‌های بعدها نسبت به هم هستند.
<br>
-ماتریس کوواریانس یک ماتریس مربعی متقارن است؛ بنابراین بردارهای ویژه آن، دو به دو بر هم عمودند.
<br>
اگر ماتریس normalized matrix را که در قسمت قبل بدست آوردیم  B بنامیم، ماتریس کوواریانس را می‌توان با فرمول زیر به دست آورد:
    </font>
</div>


<font size=4>
$S = \frac{1}{N-1}  {B}{B}^{T}$
    </font>

<div dir="rtl", style="line-height:300%;">
    <font face="XB Zar">
        <h3>پیدا کردن بردارهای ویژه و مقادیر ویژه</h3>
<font face="XB Zar" size=4>
<strong><title>قسمت سوم</title></strong>
<br>
در این قسمت باید بردارهای ویژه و مقادیر ویژه ماتریس کواریانس را حساب کنید.
        برای اینکار می‌توانید از توابع آماده‌ی نامپای استفاده کنید.
        </font>
    </font>
</div>


```python
# Calculate eigenvalues and eigenvectors here
def cov_calc(B):
    print(len(B))
    return 1/(len(B.T) - 1) * np.dot(B, B.T)
cov = cov_calc(B.T)
cov.shape
```

```python
print(cov)
```

```python
cov_by_np = np.cov(B.T)
cov_by_np = np.round(cov_by_np, 2)
print(cov_by_np)
```

همانطور که مشاهده می‌شود, نتیجه به دست امده با نتیجه به‌دست آمده از نامپای یکی است

```python
def eig_ret(cov):
    return np.linalg.eig(cov)
eig_values, eig_vectors = eig_ret(cov)
```

```python
print("eigen values are \n",eig_values )
print()
print("eigen vectors are \n",eig_vectors )
```

 مقادیر ویژه را در دو نمودار میله‌ای به ترتیب شامل خود مقادیر  ویژه و شامل نسبت هر مقدار ویژه به جمع کل آنها،
    مقایسه کنید.

```python
def data_loader(filename):
    image = Image.open(filename)
    width, height = image.size
    npimage = np.array(image)
    arr = []
    for y in range(height - 1):
        for x in range(width - 1):
            arr.append(npimage[y, x])

    nparr = np.array(arr)
    return nparr, width, height
```

```python
def data_save(new_matrix, file_name, width, height):
    image_size=(height-1)*(width-1)
    zero_matrix = np.zeros([image_size, 1])
    hstack_matrix=np.hstack((new_matrix, zero_matrix))
    new_image = Image.fromarray(np.reshape(hstack_matrix, (height-1, width-1, 3)).astype('uint8'))
    new_image.save(file_name)
```

```python
import numpy as np
from sklearn.datasets import fetch_openml
from sklearn.preprocessing import MinMaxScaler
from sklearn.decomposition import PCA, KernelPCA

def plot_digits(X, title):
    """Small helper function to plot 100 digits."""
    fig, axs = plt.subplots(nrows=10, ncols=10, figsize=(8, 8))
    for img, ax in zip(X, axs.ravel()):
        ax.imshow(img.reshape((16, 16)), cmap="Greys")
        ax.axis("off")
    fig.suptitle(title, fontsize=24)

def noise(X):
    """Usage: X_noisy = X + noise"""
    noise = np.random.normal(scale=0.25, size=X.shape)
    return noise



X, y = fetch_openml(data_id=41082, as_frame=False, return_X_y=True)
X = MinMaxScaler().fit_transform(X)
```

```python
plot_digits(X, "1")
```

```python
# Create a noisy dataset

# Perform PCA and kernelPCA on the noisy dataset
# Reconstruct images and analyze denoising potential of PCA and kernelPCA
```

```python
X_noisy = X + noise(X)
plot_digits(X_noisy, "1_noisy")
```

```python
pca = PCA(n_components=1)
pca.fit(X_noisy)
denoise_pca = pca.inverse_transform(pca.transform(X_noisy))
plot_digits(denoise_pca, "1_pca")
```

```python
pca = PCA(n_components=30)
pca.fit(X_noisy)
denoise_pca = pca.inverse_transform(pca.transform(X_noisy))
plot_digits(denoise_pca, "30_pca")
```

```python
kernel_pca = KernelPCA(
    n_components=32, gamma=1e-4, fit_inverse_transform=True
)
fited = kernel_pca.fit(X_noisy)
X_kernel_pca = kernel_pca.inverse_transform(
    kernel_pca.transform(X_noisy)
)
plot_digits(X_kernel_pca,"kernel_pca")
```
